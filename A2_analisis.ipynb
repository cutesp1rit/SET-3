```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'A2_results.txt'
data = pd.read_csv(file_path, sep=" ", header=None)
```

```python
data.columns = [
    'MergeSort_Random',
    'MergeSort_Reversed',
    'MergeSort_AlmostSorted',
    'MergeInsertion_Random',
    'MergeInsertion_Reversed',
    'MergeInsertion_AlmostSorted'
]

def plot_results(data, title, columns, labels, ylabel):
    plt.figure(figsize=(10, 6))
    for col, label in zip(columns, labels):
        plt.plot(data.index, data[col], label=label)
    plt.title(title)
    plt.xlabel('N')
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()

data.index = range(500, 10100, 100)
```

```python
plot_results(
    data,
    "Performance of Merge Sort",
    ['MergeSort_Random', 'MergeSort_Reversed', 'MergeSort_AlmostSorted'],
    ['Random Arrays', 'Reversed Arrays', 'Almost Sorted Arrays'],
    'Time (seconds)'
)
```

```python
plot_results(
    data,
    "Performance of Merge + Insertion Sort",
    ['MergeInsertion_Random', 'MergeInsertion_Reversed', 'MergeInsertion_AlmostSorted'],
    ['Random Arrays', 'Reversed Arrays', 'Almost Sorted Arrays'],
    'Time (seconds)'
)
```

```python
plot_results(
    data,
    "Comparison on Random Arrays",
    ['MergeSort_Random', 'MergeInsertion_Random'],
    ['Merge Sort', 'Merge + Insertion Sort'],
    'Time (seconds)'
)
```

```python
plot_results(
    data,
    "Comparison on Reversed Arrays",
    ['MergeSort_Reversed', 'MergeInsertion_Reversed'],
    ['Merge Sort', 'Merge + Insertion Sort'],
    'Time (seconds)'
)
```

```python
plot_results(
    data,
    "Comparison on Almost Sorted Arrays",
    ['MergeSort_AlmostSorted', 'MergeInsertion_AlmostSorted'],
    ['Merge Sort', 'Merge + Insertion Sort'],
    'Time (seconds)'
)
```
